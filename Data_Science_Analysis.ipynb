# **Day 3: Top Job Titles**
● Analyze and visualize most in-demand job titles
● Group similar titles using NLP or string match ("Data Analyst", "Analyst - Data")
Output: Bar plot or pie chart of top 10 job titles

```python
# defining a map of job titles to standardize the categories
job_map = {
    'Data Scientist': ['data scientist', 'sr. data scientist', 'senior data scientist', 'staff data scientist', 'lead data scientist', 'jr. data scientist', 'principal data scientist', 'data scientist sr', 'data scientist ii'],
    'Data Analyst': ['data analyst', 'junior data analyst', 'sr. data analyst', 'senior data analyst', 'lead data analyst', 'business data analyst', 'analytics consultant'],
    'Machine Learning Engineer': ['machine learning', 'ml engineer', 'ml researcher', 'machine learning engineer', 'machine learning scientist'],
    'Data Engineer': ['data engineer', 'etl engineer', 'big data engineer', 'data modeler', 'data pipeline', 'data architect'],
    'Business Analyst': ['business analyst'],
    'Research Scientist': ['research scientist', 'senior research scientist', 'scientist', 'principal scientist', 'staff scientist'],
    'AI/ML Scientist': ['ai', 'artificial intelligence'],
    'Software Engineer': ['software engineer', 'software developer', 'data software'],
    'Clinical Scientist': ['clinical data', 'clinical laboratory scientist', 'medical lab scientist'],
    'Other': []
}
```

```python
# Function to standardize jobs based on the job map
def standardize_job_title(raw_title):
  raw_title=str(raw_title).lower()
  for clean_title, keywords in job_map.items():
    for keyword in keywords:
      if keyword in raw_title:
        return clean_title
  return 'Other'
```

```python
# applyong the function to job title column
df['Cleaned Title']=df['Job Title'].apply(standardize_job_title)
```

```python
#plotting a bar chart for top 10 categories of job
plt.figure(figsize=(10,6))
top_titles=df['Cleaned Title'].value_counts().nlargest(10)
sns.barplot(y=top_titles.index,x=top_titles.values)
plt.title("Top 10 Job Categories")
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

# **Day 6: Skill Extraction (Text Mining)**
● Tokenize and clean job description text
● Extract skills using keyword matching or NLP (use spaCy, or predefined skill list)
● Separate soft vs technical skills (bonus)
Output: WordCloud or bar chart of most common skills

```python
#importing regex and counter for natural language processing
import re
from collections import Counter
```

```python
# defining technical and soft skills
technical_skills=['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau','data visualization', 'pandas', 'numpy', 'excel', 'git','big_data','sql','power bi', 'tensorflow','pytorch', 'machine learning', 'deep learning', 'nlp']
soft_skills=['communication','teamwork','problem solving','innovative','curious','collaboration','adaptability']
```

```python
# function to extract the skills from the job description
def skill_extract(describe,skill):
  describe=str(describe).lower()
  true_skills=[]
  for s in skill:
    if re.search(r'\b'+re.escape(s)+r'\b',describe):
      true_skills.append(s)
  return true_skills
```

```python
#applying the function to segregate the techincal skills and soft skills from the job description
df['Technical Skills']=df['Job Description'].apply(lambda x:skill_extract(x,technical_skills))
df['Soft Skills']=df['Job Description'].apply(lambda x:skill_extract(x,soft_skills))
```

```python
# sum of the already mentioned skills in the dataset and the extracted ones
already_skills=['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau','big_data']
techi_skills=df[already_skills].sum().to_dict()
```

```python
techi_skills
```

```python
#counting additional technical skills retrieved from the job description
extra_skills=Counter()
for skill in df['Technical Skills']:
  for s in skill:
    extra_skills[s]=extra_skills[s]+1
```

```python
#counting additional soft skills obtained from the job description
extra_sof_skill=Counter()
for skill in df['Soft Skills']:
  for s in skill:
    extra_sof_skill[s]=extra_sof_skill[s]+1
```

```python
#combining pre defined and extracted technical skills
complete_tech=techi_skills.copy()
for skill in extra_skills:
  if s in complete_tech:
    complete_tech[skill]=max(complete_tech[skill],extra_skills[skill])
  else:
    complete_tech[skill]=extra_skills[skill]
```

```python
# updating the combine variable to be the mix of techincal and soft skills
combine=complete_tech.copy()
combine.update(extra_sof_skill)
```

```python
combine
```

```python
# creating a data frame of all skills with their counts
df_skill=pd.DataFrame(list(combine.items()),columns=['Skill','Count'])
df_skill
```

```python
# after sorting the skills plotting the top 10 highly demanding technical skills
df_skill=df_skill.sort_values('Count',ascending=False)
plt.figure(figsize=(10,4))
sns.barplot(x='Count',y='Skill',data=df_skill.head(10))
plt.title('Top 10 Skills')
plt.xlabel('Count')
plt.ylabel('Skill')
plt.show()
```

```python
#sorting the soft skills and creating a dataframe of softskills with the count
df_soft=pd.DataFrame(list(extra_sof_skill.items()),columns=['Skill','Count'])
df_soft=df_soft.sort_values('Count',ascending=False)
```

```python
#plotting the top 10 soft skills
plt.figure(figsize=(10,4))
sns.barplot(x='Count',y='Skill',data=df_soft.head(10))
plt.title('TOP 10 SOFT SKILLS')
plt.xlabel('COUNT')
plt.ylabel('SKILL')
```

# **Day 7: Skill Trends Across Job Levels**
● Segment job postings by level (e.g., entry-level, senior)
● Compare required skills per level
● Use visual comparisons
Output: Comparative skill charts for each job level

```python
#function to categorise the carrer based on job titles
def organise(titles):
  titles=str(titles).lower()
  if 'senior' in titles or 'sr.' in titles or 'lead' in titles or 'principal' in titles:
    return 'Senior'
  elif 'junior' in titles or 'jr.' in titles or 'entry' in titles:
    return 'Entry'
  else:
    return 'Mid'
```

```python
# applying the function to get carrer level categorisation
df['Career']=df['Cleaned Title'].apply(organise)
```

```python
# getting unique career levels
vacancy_level=df['Career'].unique().tolist()
```

```python
vacancy_level
```

```python
# creating a list of already defined techical skills
must_skills=['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big_data']
```

```python
# sumation of skills with carrer level
skill_aggregate=df.groupby('Career')[must_skills].sum()
```

```python
# aggregating the already defined , extracted skills with career level
tier_skill={}
for tier in vacancy_level:
  tier_df = df[df['Career']==tier]
  tech_total=Counter()
  for skill in must_skills:
    tech_total[skill]=tier_df[skill].sum()
  tier_skill[tier]=tech_total
```

```python
#extracting additional technical and soft skills by career level
for tier in vacancy_level:
  tier_df=df[df['Career']==tier]
  tech_add=Counter()
  soft_add=Counter()
  for skills in tier_df['Technical Skills']:
    if isinstance(skills,str):
      skills=eval(skills)
    for skill in skills:
      tech_add[skill]+=1
  for skills in tier_df['Soft Skills']:
    if isinstance(skills,str):
      skills=eval(skills)
    for skill in skills:
      soft_add[skill]+=1
  tier_skill[tier].update(tech_add)
  tier_skill[tier].update(soft_add)
```

```python
#combining all skills for career level
all_skills=Counter()
for tier in tier_skill:
  all_skills.update(tier_skill[tier])
```

```python
#retreiving the top 5 skills
top_skill=[skill for skill,count in all_skills.most_common(5)]
```

```python
#creating a dataframe for plotting skills
plot_1=pd.DataFrame(index=vacancy_level,columns=top_skill)
```

```python
#the plot of skills with the career level
for tier in tier_skill:
  for skill in top_skill:
    plot_1.loc[tier,skill]=tier_skill[tier].get(skill,0)
```

```python
#plotting a bar chart of top skills by career level
plot_1.plot(kind='bar',figsize=(12,6))
plt.title('Top Skills by Career Level')
plt.xlabel('Career Level')
plt.ylabel('Number of Jobs')
plt.show()
```

# **Day 8: Skill Demand Across Companies/Industries**
● Explore how skill requirements vary by company or industry
● Group and compare most requested skills across categories
Output: Multi-category comparison plot

```python
#aggregating the pre defined skills by sector of the industry
must_skills
industry_skill=df.groupby('Sector')[must_skills].sum()
```

```python
# combing the techincal and soft skills by industry sector
industry_ability={}
for sector in df['Sector'].unique():
  jobs=df[df['Sector']==sector]
  tech_counts=Counter()
  for skills in jobs['Technical Skills']:
    if isinstance(skills,str):
      skills=eval(skills)
    for skill in skills:
      tech_counts[skill]+=1
      soft_counts=Counter()
  for skills in jobs['Soft Skills']:
    if isinstance(skills,str):
      skills=eval(skills)
    for skill in skills:
      soft_counts[skill]+=1
  combined_abilities=industry_ability.get(sector,{}).copy()
  combined_abilities.update(tech_counts)
  combined_abilities.update(soft_counts)
  industry_ability[sector]=combined_abilities
```

```python
# selecting the top 5 industries by postings
top_industry=df['Sector'].value_counts().head(5).index
```

```python
# defining the significant technical and soft skills for plot
key_skills=['python', 'excel', 'hadoop', 'spark', 'aws']
key_soft=['communication','innovative']
```

```python
#creating a dataframe for visualising the skill distribution
visually_data=industry_skill.loc[top_industry,key_skills].copy()
for skill in key_soft:
  visually_data[skill]=[industry_ability[sector].get(skill,0) for sector in top_industry]
```

```python
# plotting skill distribution across top industries
visually_data.plot(kind='bar',figsize=(12,6))
plt.title('Skill Distribution Across Top Industries')
plt.xlabel('Industry Sector')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45,ha='right')
plt.legend(title='Skills')
plt.tight_layout()
plt.show()
```
